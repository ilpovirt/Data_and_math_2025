```python
#Let's write normal distribution as a function in Python
#Several libraries include it but let's do this for practise
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x,mu,sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-1/2*((x-mu)/sigma)**2)

#Range of values
x = np.arange(-5,5,0.01)
mu = 0
sigma = 1 #Standard normal distrubution

plt.figure(figsize=(10,4))
plt.plot(x,gaussian(x,mu,sigma))
plt.axis([-4,4,0,0.45])
plt.grid()
plt.show()
```

```python
# 42. The continuous random variable x has a standard normal distribution. Calculate the fraction of values x what fullfills the condition:

# a) x  <  0
# b) x  >  1.2
# c) x  >  -1.2
# d) x  <  -3

#Method 1: Own function and numerical integration
def gaussian_standard(x):
    mu = 0
    sigma = 1
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-1/2*((x-mu)/sigma)**2)

#Method 1.1, numerical integration usign summing function
#Define the range of integral:
dx = 0.001
x = np.arange(1.2,999,dx) #999 is large enough, but not infinity
I = np.sum(gaussian_standard(x)*dx)
print('The fraction of values of x larger than 1.2 is ', I)

#Method 1.2
import scipy
I = scipy.integrate.quad(gaussian_standard,1.2,np.inf)
print('The fraction of values of x larger than 1.2 is ', I[0])

#Method 2
#Use cumulative distribution function (CDF)
#The value of CDF gives the integral of distribution function from minus infinity to x

1 - scipy.stats.norm.cdf(1.2) #1 minus integral from -inf to 1.2

print('\n Task 42b: The fraction of values with x > 1.2 is', 1 - scipy.stats.norm.cdf(1.2) )
print('Task 42c: The fraction of values with x > - 1.2 is',1 - scipy.stats.norm.cdf(-1.2) )
print('Task 42d: The fraction of values with x < -3 is',scipy.stats.norm.cdf(-3))
```
