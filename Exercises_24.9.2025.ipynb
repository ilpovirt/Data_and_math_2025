```python
#Let's make a graph of function and its derivative
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5,4,0.01) #Define the range of x
y = x**2 - x - 6 #Define the value of function for all the values of x
dy = 2*x-1 #derivative
plt.plot(x,y)
plt.plot(x,dy)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','Derivative of f(x)'])
plt.axis([-4,4,-7,6])
plt.grid()
plt.axhline(linewidth = 2, color = 'k')
plt.axvline(linewidth = 2, color = 'k')
```

```python
#b) Calculate the derivative of f(x) analytically using Python (Sympy)  
import sympy as sm 
#Define symbolic independent variable
x = sm.Symbol('x')
#Define a function (Not like Python function!)
f = x**2 - x - 6
#Calculate the derivative
df = f.diff(x)
print('The derivative is:' )
df
```

```python
#c) Calculate the derivative of f(x) numerically using Python 
#Define function f(x)

f = x**2 - x -6
#Define the values of x 
x = np.arange(-5,4,0.01)

#Calculate the derivative. Gradient refers to rate of change also in case of multidimensional functions. 
#In one dimension gradient is practically derivative. 
df = np.gradient(f,x)

#Let's see if it worked...
plt.plot(x,f)
plt.plot(x,df)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','Derivative of f(x)'])
plt.axis([-4,4,-7,6])
plt.grid()
plt.axhline(linewidth = 2, color = 'k')
plt.axvline(linewidth = 2, color = 'k')
```

```python
#g) Calculate the area enclosed by f(x) and x-axis using Python. 
# Write your own code according to idea of calculating the area. 
#Define the function
def f(x):
    return x**2-x-6
#Width of a stripe
dx = 0.00001
#Define the values of x
x = np.arange(-2,3,dx)
#Calculate area (value of define integral)
#Area of a single stripe is f(x)*dx
#--> Total area is the sum of stripes
A = np.sum(f(x)*dx)
print('The value of integral is ', A)
A_true = (1/3*3**3-1/2*3**2-6*3) - (1/3*(-2)**3-1/2*(-2)**2-6*(-2))
print('The difference between numerical solution and true value (residual) is ', A_true - A )
```
