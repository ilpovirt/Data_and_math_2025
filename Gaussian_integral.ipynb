```python
import matplotlib.pyplot as plt
import numpy as np
def gaussian(x, mu, sigma):
    return 1./(np.sqrt(2.*np.pi)*sigma)*np.exp(-np.power((x - mu)/sigma, 2.)/2)

def Gaussian_integral(mu,sigma,lim1,lim2):
    t=0.001*sigma
    xlim1 = -4*sigma+mu
    xlim2 = 4*sigma+mu
    x = np.arange(xlim1,xlim2,t)
    y = gaussian(x, mu, sigma)
    plt.plot(x, y, label='Gaussian', color='blue')
    x2 = np.arange(lim1,lim2 ,t)
    y2 = gaussian(x2, mu, sigma)
    plt.fill_between(x2, y2, 0, color='skyblue', alpha=0.3)
    plt.plot([lim1,lim1],[0,gaussian(lim1, mu, sigma)],'k')
    plt.plot([lim2,lim2],[0,gaussian(lim2, mu, sigma)],'k')
    # Set labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    value = np.sum(gaussian(x2, mu, sigma)*t)
    otsikko =  'The integral of x from '+str(lim1)+ ' to '+ str(lim2) + ' is ' + str(np.round(value,4))
    plt.title(otsikko)
    plt.legend()
    plt.axis([xlim1,xlim2,0,np.max(y)*1.04])
    # Display the plot
    plt.show()

mu = 0
sigma = 1
lim1 = 2*sigma+mu
lim2 = 2*sigma+mu
Gaussian_integral(mu,sigma,lim1,lim2)
```
