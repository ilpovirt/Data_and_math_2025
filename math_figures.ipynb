```python
import numpy as np
import matplotlib.pyplot as plt

#number generator and histogram plotter

def gaussian(x, mu, sigma):
    return 1./(np.sqrt(2.*np.pi)*sigma)*np.exp(-np.power((x - mu)/sigma, 2.)/2)


mu = 0
sigma = 1
n = 10000
s = np.random.normal(mu, sigma, n)
t=0.5
x = np.arange(-4+mu,4+mu,t)

#print('Random values are ',s)
fig  = plt.figure(figsize=(25,5), dpi = 100)
plt.subplot(1,3,1)
plt.hist(s, bins = x, width = 0.49)
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('Count')
plt.grid(True)

t=0.1
x = np.arange(-4+mu,4+mu,t)

plt.subplot(1,3,3)
plt.plot(x,gaussian(x, mu, sigma)) 
plt.xlabel('x')
plt.ylabel('')
plt.grid(True)
#plt.legend(['Distribution function','Histogram'])
plt.title('Normalized distribution function')

plt.subplot(1,3,2)
plt.plot(x,gaussian(x, mu, sigma)*n) 
plt.xlabel('x')
plt.ylabel('')
plt.grid(True)
#plt.legend(['Distribution function','Histogram'])
plt.title('Distribition function')
plt.show()

```

```python
mu = 0
sigma = 1
n = 100000
s = np.random.normal(mu, sigma, n)
t=0.5
x = np.arange(-4+mu,4+mu,t)

#print('Random values are ',s)
fig  = plt.figure(figsize=(25,5), dpi = 100)
plt.subplot(1,3,1)
plt.hist(s, bins = x, width = 0.49)
plt.title('Histogram')
plt.xlabel('x')
plt.ylabel('Count')
plt.grid(True)
plt.axis([0,2,0,20000])

t=0.1
x = np.arange(-4+mu,4+mu,t)

plt.subplot(1,3,3)
plt.plot(x,gaussian(x, mu, sigma)) 
plt.xlabel('x')
plt.ylabel('')
plt.grid(True)
#plt.legend(['Distribution function','Histogram'])
plt.title('Normalized distribution function')
np.sum(gaussian(x, mu, sigma)*t)
plt.axis([0,2,0,1])

plt.subplot(1,3,2)
plt.plot(x,gaussian(x, mu, sigma)*10000) 
plt.xlabel('x')
plt.ylabel('')
plt.grid(True)
#plt.legend(['Distribution function','Histogram'])
plt.title('Distribition function')
np.sum(gaussian(x, mu, sigma)*t)
plt.axis([0,2,0,4000])
plt.show()
```
